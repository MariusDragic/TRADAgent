```python
import requests
import pandas as pd
import os
```

```python
API_KEY = os.getenv("ALPHAVANTAGE_API_KEY")
url = f"https://www.alphavantage.co/query?function=NATURAL_GAS&interval=daily&apikey={API_KEY}"
r = requests.get(url)
data = r.json()



```

```python
data
```

```python
import pandas as pd

df = pd.DataFrame(data["data"])

df["date"] = pd.to_datetime(df["date"])

df["value"] = pd.to_numeric(df["value"], errors="coerce")

# Supprimer les jours non cots
df = df.dropna(subset=["value"])

# Trier par date
df = df.sort_values("date").reset_index(drop=True)

# Sauvegarde CSV
df.to_csv("natural_gas_daily_clean.csv", index=False)

print(" CSV saved: natural_gas_daily_clean.csv")
print(df.head())
```

```python
import os
import json
import time
import requests
from datetime import datetime
from pathlib import Path

# ==========================
# CONFIG
# ==========================

url = f"""https://www.alphavantage.co/query?function=NEWS_SENTIMENT&topics=blockchain&sort=RELEVANCE&limit=1000&apikey={API_KEY}"""

data = requests.get(url).json()

DATA_DIR = Path("data/news/macro")
DATA_DIR.mkdir(parents=True, exist_ok=True)

# ==========================
# FETCH FUNCTION
# ==========================

def save_json(data: dict) -> Path:
    date_str = datetime.utcnow().strftime("%Y-%m-%d")
    file_path = DATA_DIR / f"news_blockchain.json"

    with open(file_path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False)

    return file_path

save_json(data)
```
